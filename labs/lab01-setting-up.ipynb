## Krok 1: Inštalácia Anacondy

Najnovšiu verziu Anacondy (s Pythonom 3.12) [nájdete tu](https://www.anaconda.com/download). Vyberte si inštalačný súbor pre váš operačný systém a postupujte podľa inštrukcií inštalácie. [Podrobné návody pre jednotlivé operačné systémy nájdete tu](https://docs.anaconda.com/anaconda/install/). Odporúčame počas inštalácie zaregistrovať Anacondu ako defaultnú inštaláciu Pythonu. Na OS Windows tiež je výhodnejšie pridať cestu k inštalácii Anacondy do zoznamu systémových premenných (ako cestu uveďte inštaláciu Anacondy, ktoré môžete nastaviť počas inštalácie - tento priečinok by mal obsahovať súbor `python.exe`).

<img src="sources/lab01/anaconda_installation.jpg" width="400">

Po ukončení inštalácie jej úspešnosť môžete skontrolovať cez príkazový riadok (alebo terminál) zadaním príkazu `python --version`. Mal by sa zobraziť výpis typu `Python 3.12.X`. Anaconda takisto má vlastný príkazový riadok Anaconda Prompt, kde môžete interagovať priamo s inštaláciou.

Anaconda ďalej umožňuje vytvorenie virtuálnych prostredí, kde môžete mať rôzne verzie Pythonu, nainštalovať si knižnice a upraviť nastavenia prostredia bez toho, aby ste zmenili samotnú inštaláciu Pythonu. Podrobný návod k virtuálnym prostrediam je dostupný na [web stránke Anacondy](https://docs.conda.io/projects/conda/en/latest/user-guide/tasks/manage-environments.html).

Ak počas semestra alebo pri riešení zadaní a vlastných projektov potrebujete nainštalovať nové knižnice, môžete tak urobiť zadaním príkazu `pip install NAZOV` alebo `conda install NAZOV` v príkazovom riadku, resp. v Anaconda Prompte.

## Krok 2: Inštalácia vývojového prostredia (IDE)

Na internete nájdete veľký počet vývojových prostredí s podporou Pythonu, z nich si vyberte podľa vlastných skúseností a preferencií. Samotná Anaconda obsahuje takéto prostredie s menom Spyder. Často používane IDE pre Python je [PyCharm](https://www.jetbrains.com/pycharm/), ktorý je dostupný zadarmo v Community Version. Jedno z viacplatformových prostredí pre podporu rôznych jazykov je [Sublime Text](https://www.sublimetext.com). Sublime Text môžete používať bezplatne, avšak po niekoľkých uloženiach editovaného súboru vám vyskočí dialógové okno s ponukou nákupu plnej verzie. Anaconda ponúka rozšírenie pre Sublime Text pre editovanie pythonovských súborov, balík Anaconda. [Inštalačný manuál nájdete tu](https://damnwidget.github.io/anaconda/).

```python
class Product:
    def __init__(self):
        self.price = 1000
        # self.name = "abc"
        # self.count = 10

    def set_price(self, new_price):
        self.price = new_price

    def sell(self):
        print(f"Sold for {self.price}")
```

```python
p1 = Product()
p1.sell()
```

```python
p1.sell()
p1.set_price(2000)
p1.sell()
```

```python
class HashTable:
    def __init__(self):
        self.table = {}

    def hash_function(self, number):
        return number % 5

    def add_number(self, number):
        key = self.hash_function(number)
        if key in self.table:
            self.table[key].append(number)
        else:
            self.table[key] = [number]
```

```python
class HashTable:
    def __init__(self):
        self.table = [[], [], [], [], []]

    def hash_function(self, number):
        return number % 5

    def add_number(self, number):
        key = self.hash_function(number)
        self.table[key].append(number)

    def print_table(self):
        print(self.table)
```

```python
h_table = HashTable()
h_table.print_table()
h_table.add_number(8)
h_table.add_number(9)
h_table.add_number(0)
h_table.add_number(4)
h_table.print_table()  # best practice
print(h_table.table)  # don't do this
```

```python
h_table.table = []
print(h_table.table)
```

```python
class BankAccount:
    def __init__(self, name, amount):
        self.name = name
        # self.amount = amount  # public variable
        self.__amount = amount

    def get_amount(self):
        return self.__amount

    def set_amount(self, new_amount):
        self.__amount = new_amount
```

```python
a1 = BankAccount("Janko Hrasko", 1000)
print(a1.get_amount())
```

```python
a1.__amount = 0
print(a1.get_amount())
```

```python
a1.set_amount(0)
print(a1.get_amount())
```

```python
class Person:
    def __init__(self, name, age):
        print("setting values in Person")
        self.name = name
        self.age = age

    def get_name(self):
        return self.name

    def get_age(self):
        return self.age
```

```python
person1 = Person("John Doe", 35)
print(person1.get_name())
```

```python
class Student:
    def __init__(self, name, age):
        print("setting values in Student")
        self.name = name
        self.age = age
        self.year = 1

    def get_name(self):
        return self.name

    def get_age(self):
        return self.age

    def get_year(self):
        return self.year

    def go_to_next_year(self):
        self.year += 1
```

```python
student = Student("Jane Doe", 19)
print(student.get_name())
print(student.get_age())
print(student.get_year())
```

```python
class Student(Person):
    def __init__(self, name, age):
        super().__init__(name, age)
        print("setting values in Student")
        self.year = 1

    def get_year(self):
        return self.year

    def go_to_next_year(self):
        self.year += 1
```

```python
student2 = Student("Janko Hrasko", 19)
print(student2.get_name())
print(student2.get_age())
print(student2.get_year())

person2 = Person("John Doe", 30)
print(person2.get_name())
print(person2.get_age())
print(person2.get_year())
```

```python
class Faculty(Person):
    def __init__(self, name, age, position):
        super().__init__(name, age)
        self.position = position

    def get_position(self):
        return self.position

    def print_info(self):
        print(f"{self.name} is working on position {self.position}")
```

```python
teacher = Faculty("John Smith", 45, "associate")
teacher.print_info()
print(teacher.get_name())
print(teacher.get_age())
print(teacher.get_year())
```

```python
people = [
    Student("X Y", 20),
    Student("A B", 21),
    Faculty("J K", 35, "assistant"),
    Student("M O", 23)
]
for person in people:
    print(person.get_name())
```
